## Web scraping para extração de planilhas de cadastros.
- O intuito desse web scrapping foi de automatizar um processo de extração de cadastro planilhas de cadastros de um sistema legado web.
- Toda migração de dados, a qual clientes utilizem esse sistema é necessário a extração de cadastros de forma manual e repetitivo.<br>
- Após extrair as planilhas de cadastros é possível formatar os dados e posteriormente migrar para um banco de dados (Mysql) os dados já formatados <br>


### Processo Webscrapping 
- Acessar site web
- Logar com usuário e senha do cliente 
- Se logar no sistema web 
- Acessar abas de download de clientes 
- fazer download 
- Salvar na máquina local 

Bibliotecas utilizadas: Selenium e Pandas.





```python
# Importando bibliotecas 
from selenium import webdriver
import pandas as pd
import time
from selenium.webdriver.common.by import By
from acesso import email,senha

# Configurando variáveis de ambiente e utilização do webscrapping em modo background
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("headless")
chrome_options.add_argument("no-sandbox")
chrome_options.add_argument("disable-dev-shm-usage")

chrome_prefs = {"download.default_directory": r"C:\TESTE_GENDO"}
chrome_options.experimental_options["prefs"] = chrome_prefs
```

```python
# Função e código web scrapping 
def gendo_cliente():
    driver = webdriver.Chrome(options=chrome_options)
    driver.get('https://acesso.gendo.app') 
    print(driver.title)
    elementos_nome = driver.find_elements(By.NAME, "email")
    elementos_nome[0].send_keys(email)
    print('E-mail preenchido com sucesso')
    elementos_senha = driver.find_elements(By.NAME, "password")
    elementos_senha[0].send_keys(senha)
    print('Senha prennchida com sucesso')
    elementos_logar = driver.find_elements(By.TAG_NAME, "button")
    print(len(elementos_logar))
    elementos_logar[0].click()
    print('Login efetuado com sucesso')
    time.sleep(3)
    elementos_clientes = driver.find_elements(By.CLASS_NAME, "link-menu")
    print(len(elementos_clientes))
    elementos_clientes[3].click()
    print('menu selecionado')
    listagem = driver.find_elements(By.PARTIAL_LINK_TEXT, "Listagem")
    print(len(listagem))
    listagem[0].click()
    print('Listagem (clientes)')
    todos_clientes = driver.find_elements(By.XPATH, '//*[@id="menu-principal"]/li[3]/ul/li[1]/a/span')
    print(len(todos_clientes))
    todos_clientes[0].click()
    print('Todos os clientes OK')
    time.sleep(1)
    todos_clientes = driver.find_elements(By.XPATH, '//*[@id="DataTables_Table_0_length"]/label/select')
    print(len(todos_clientes))
    todos_clientes[0].click()
    todos_clientes[0].send_keys('5000')
    print('5000 clientes selecionados')
    time.sleep(5)
    download = driver.find_elements(By.XPATH, '//*[@id="toolbar"]/form/div/button[4]/i')
    print(len(download))
    download[0].click()
    time.sleep(10)
    return print('Dowload realizado com sucesso !!!')
```

```python
# Chamando a função para realizar o web scrapping e fazer o download do arquivo
gendo_cliente()
```
